```python
#fillna is use for fill all null values
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
df["country"] = df["country"].fillna("Unknown")
```

```python
movies = df[df['type'] == 'Movie'].copy()

movies['duration'] = movies['duration'].str.replace(' min', '')
movies['duration'] = pd.to_numeric(movies['duration'])

movies['duration'].describe()
```

```python
plt.figure(figsize=(10,5))
plt.hist(movies['duration'], bins=30)
plt.title("Movie Duration Distribution")
plt.xlabel("Minutes")
plt.ylabel("Count")
plt.show()
```

```python
top_directors = df['director'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_directors.plot(kind='bar')
plt.title("Top 10 Directors")
plt.show()
```

```python
genres = df['listed_in'].str.split(', ')
genres = genres.explode()

top_genres = genres.value_counts().head(10)

plt.figure(figsize=(10,5))
top_genres.plot(kind='bar')
plt.title("Top Genres on Netflix")
plt.show()
```

# Netflix Data Analysis Project

## Project Overview

This project analyzes the Netflix Movies and TV Shows dataset using Python and data analysis techniques. The objective was to explore Netflix's content library, identify trends, and extract meaningful business insights through data cleaning, exploratory data analysis (EDA), and visualization.

## Tools and Technologies Used

* Python
* Pandas
* NumPy
* Matplotlib
* Jupyter Notebook

## Project Workflow

### 1. Data Loading

The Netflix dataset was imported into Pandas DataFrames for analysis.

### 2. Data Cleaning

* Checked for missing values.
* Handled null values using appropriate techniques.
* Removed duplicate records.
* Converted date columns into datetime format.
* Standardized data formats for analysis.

### 3. Exploratory Data Analysis (EDA)

#### Content Type Analysis

Compared the number of Movies and TV Shows available on Netflix.

**Finding:**
Movies are more numerous than TV Shows on Netflix.

#### Genre Analysis

Analyzed the most common content genres.

**Finding:**
International Movies is the most common genre on Netflix.

#### Country Analysis

Examined which countries contribute the most content.

**Finding:**
The United States contributes the highest number of titles.

#### Rating Analysis

Studied the distribution of content ratings.

**Finding:**
TV-MA is the most common content rating, indicating a large amount of mature-audience content.

#### Duration Analysis

Analyzed movie duration trends.

**Finding:**
The average movie duration is approximately 99.5 minutes.

#### Content Addition Trend

Investigated how Netflix content grew over time.

**Finding:**
Netflix added the highest amount of content in 2019.

## Key Insights

1. Movies dominate Netflix's content catalog.
2. International Movies are highly popular on the platform.
3. The United States is the leading content-producing country.
4. Netflix contains a significant amount of mature-audience content.
5. Most movies have a duration close to 100 minutes.
6. Netflix experienced major content expansion in 2019.
